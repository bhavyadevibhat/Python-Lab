```python
def acircle(r):
    return 3.14*r*r
def pcircle(r):
    return 3.14*2*r
def arectangle(l,b):
    return l*b
def prectangle(l,b):
    return 2*(l+b)
def atriangle(b,h):
    return 0.5*b*h
def ptriangle(b,s1,s2):
    return b+s1+s2
while(1):
    print("1.Area and Perimeter of Circle\n2.Area and Perimeter of rectangle\n3.Area and Perimeter triangle\n4.exit")
    choice=int(input("Enter your choice:"))
    if choice==1:
        radius=float(input("Enter Radius:"))
        print("Area of  circle :",acircle(radius))
        print("Perimeter of circle:",pcircle(radius))
    elif choice==2:
        length=float(input("Enter length of rectangle:"))
        breadth=float(input("Enter breadth of rectangle:"))
        print("Area of rectangle is:",arectangle(length,breadth))
        print("Perimeter of rectangle :",prectangle(length,breadth))
    elif choice==3:
        base=float(input("Enter base of triangle:"))
        height=float(input("Enter height of triangle:"))
        print("Area of triangle:",atriangle(base,height))
        side1=float(input("Enter side1 of triangle:"))
        side2=float(input("Enter side2 of triangle:"))
        print("Perimeter of triangle is:",ptriangle(base,side1,side2))
    elif choice==4:
        break
    else:
        print("Invalid Choice")


```

```python
list1=[]
n=int(input("Enter Size:"))
for i in range(n):
    list1.append(int(input("Enter element:")))
print("Original List is:",list1)
while(1):
    print("1.Insert\n 2.Remove\n 3.Append\n 4.Len\n 5.Pop\n 6.Clear\n 7.Exit")
    ch=int(input("Enter your Choice:"))
    if ch==1:
        ele=int(input("Enter Element to insert"))
        index=int(input("Enter Index value:"))
        list1.insert(index,ele)
        print("After insertion",list1)
    elif ch==2:
        ele=int(input("Enter value to delete:"))
        if ele in list1:
            list1.remove(ele)
            print("After Removing the list is:",list1)
        else:
            print("Element Not found in list!")  
    elif ch==3:
        ele=int(input("Enter a value to append"))
        list1.append(ele)
        print("After appending:",list1)
    elif ch==4:
        print("The length of list is:",len(list1))
    elif ch==5:
        ele=list1.pop()
        print("The poped element is:",ele)
        print("After poping the list is:",list1)
    elif ch==6:
        list1.clear()
        print("The list is empty!",list1)
    elif ch==7:
        break
    else:
        print("invalid Input!!")
```

```python
class employee:
    def __init__(self,empid,name,designation,salary):
        self.empid=empid
        self.name=name
        self.designation=designation
        self.salary=salary
    def empdetails(self):
        print(self.empid,end="\t\t")
        print(self.name,end="\t\t")
        print(self.designation,end="\t\t")
        print(self.salary,end="\t\t")
        print(self.grosssal(),end="\t\t\t")
        print(self.netsal())
    def grosssal(self):
        DA=5/100*self.salary
        HRA=10/100*self.salary
        gross=self.salary+DA+HRA
        return gross
    def netsal(self):
        LIC=3/100*self.salary
        PF=8/100*self.salary
        deduction=LIC+PF
        netsalary=self.grosssal()-deduction
        return netsalary
empid=[]
name=[]
designation=[]
salary=[]
for i in range(5):
    empid.append(int(input("Enter Register Number:")))
    name.append(input("Enter Name:"))
    designation.append(input("Enter Designation:"))
    salary.append(float(input("Enter Basic Salary:")))

emp1=employee(empid[0],name[0],designation[0],salary[0])
emp2=employee(empid[1],name[1],designation[1],salary[1])
emp3=employee(empid[2],name[2],designation[2],salary[2])
emp4=employee(empid[3],name[3],designation[3],salary[4])
emp5=employee(empid[4],name[4],designation[4],salary[4])
print("**********************************Employee Details*****************************************************")
print("-------------------------------------------------------------------------------------------------------")
print("Emp Id:",end="\t\t")
print("Name:",end="\t\t")
print("department:",end="\t\t")
print("Salary:",end="\t\t")
print("Gross Salary:",end="\t\t")
print("Net salary:")
print("-------------------------------------------------------------------------------------------------------")
emp1.empdetails()
emp2.empdetails()
emp3.empdetails()
emp4.empdetails()
emp5.empdetails()


        
```
