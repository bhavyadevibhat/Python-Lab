```python
def add(list1):
    sum1=0
    for i in list1:
        sum1+=i
    return sum1
def sub(list1):
    diff=list1[0]
    for i in range(1,len(list1)):
        diff-=list1[i]
    return diff
def mul(list1):
    mul=1
    for i in list1:
        mul*=i
    return mul
def div(list1):
    div=list1[0]
    for i in range(1,len(list1)):
        div/=list1[i]
    return div
while(1):
    print("********************************************************************")
    print(" 1.Additon\n 2.Substraction\n 3.Multiplication\n 4.Division\n 5.Exit")
    print("********************************************************************")
    choice=int(input("Enter your choice:"))
    if choice==1:
        list1=[]
        n=int(input("How many numbers you want to add?"))
        for i in range(n):
            list1.append(float(input("Enter numbers:")))
        print("The Adittion of given number is : ",add(list1))
    elif choice==2:
        list1=[]
        n=int(input("How many numbers you want to add?"))
        for i in range(n):
            list1.append(float(input("Enter numbers:")))
        print("The Difference of given number is : ",sub(list1))
    elif choice==3:
        list1=[]
        n=int(input("How many numbers you want to add?"))
        for i in range(n):
            list1.append(float(input("Enter numbers:")))
        print("The Adittion of given number is : ",mul(list1))
    elif choice==4:
        list1=[]
        n=int(input("How many numbers you want to add?"))
        for i in range(n):
            list1.append(float(input("Enter numbers:")))
        print("The Adittion of given number is : ",div(list1))
    elif choice==5:
        break
    else:
        print("Invalid Choice")
        
    
```
